```python
import pandas as pd
import requests
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
```

# Request page source from URL

```python
url = "https://www.rottentomatoes.com/browse/movies_at_home/critics:certified_fresh,fresh~sort:popular?page=10"
```

```python
page = requests.get(url)
page
```

```python
## display the page source code
page.content
```

```python
soup = soup(page.content, "html.parser")
print(soup.prettify())
```

```python
movie_name=[]
year=[]
rating=[]
```

```python
movie_elements = soup.find_all('div', class_='js-tile-link')

for movie_element in movie_elements:
    # Extracting movie name
    movie_name.append(movie_element.find('span', class_='p--small').text.strip())

    # Extracting critic score
    rating.append(movie_element.find('score-pairs-deprecated')['criticsscore'])

    # Extracting streaming date
    year.append(movie_element.find('span', class_='smaller').text.strip().replace('Streaming ', ''))
```

```python
years = [date[-4:] for date in year]
years[0:6]
```

```python
data = {'Movie Name': movie_name, 'Critic Score': rating, 'Streaming Year': year}
df = pd.DataFrame(data)
```
